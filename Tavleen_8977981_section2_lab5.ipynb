### IMPORTING TB RAW DATASET

```python
import pandas as pd
df = pd.read_csv("C:/Users/HP/Downloads/CSCN8000/LAB5_CSCN8000/data/tb-raw.csv")
print(df.head(7))
```

### DATA TO READABLE DATA

```python
#Data Reshaping
df_tidy = pd.melt(df, id_vars=['country', 'year'], var_name='sex_age', value_name='cases') 

 #Split the variable
df_tidy['sex'] = df_tidy['sex_age'].apply(lambda x: x[0]) 
df_tidy['age'] = df_tidy['sex_age'].apply(lambda x: x[1:])  

#Replace the names
df_tidy.replace('m', 'Man', inplace=True)
df_tidy.replace('f', 'Women', inplace=True)

#Map age group labels
df_tidy['age'] = df_tidy['age'].map({
    '014': '0 to 14',
    '1524': '15 to 24',
    '2534': '25 to 34',  
    '3544': '35 to 44',
    '4554': '45 to 54',
    '5564': '55 to 64',
    '65': '65',
    'u': 'unidentical'
})

#Remove 'sex_age'
df_tidy.drop(columns=['sex_age'], inplace=True) 
print(df_tidy)
```

```python
#Checing Missing Values
missing_values = df_tidy.isnull().sum()
print(missing_values)
```

```python
#Null value in the place of 0
df_tidy = df_tidy.fillna(0)
print(df_tidy)
```

```python
# Importing
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Data URL
data_url = "http://lib.stat.cmu.edu/datasets/boston"

# Load the dataset into a pandas DataFrame
boston_data = pd.read_csv(data_url, sep='\s+', skiprows=22, header=None)

# Let's print the first few rows of the dataset to verify it has been loaded correctly
print(boston_data.head())

# Fill "NaN" with 0.
boston_data = boston_data.fillna(0)
print(boston_data)
```

### GIVING COLUMN NAMES

```python
column_m = [] # empty array
for i in range(0, len(boston_data), 2):
    merged_row = pd.concat([boston_data.iloc[i], boston_data.iloc[i+1]], ignore_index=True)
    column_m.append(merged_row)

merged_data = pd.DataFrame(column_m) 

merged_data = merged_data.iloc[:, :14] # array slicing

merged_data.columns = [
    "CRIM", "ZN", "INDUS", "CHAS", "NOX", "RM", "AGE", "DIS", "RAD", "TAX", "PTRATIO", "B", "LSTAT", "MEDV"
] 
print(merged_data) # view
```

### VISUALISATION

```python
for column in merged_data:
    plt.figure(figsize=(15, 15))  # Set the figure size
    merged_data.boxplot([column])

    plt.title('Boxplot of ' + column)  # title
    plt.ylabel('Values')  # label
    plt.show()  # Display
```

### STATISTICS

```python
import numpy as np
from scipy import stats
z_scores = np.abs(stats.zscore(merged_data))
threshold = 3
outliers_z = (z_scores > threshold).any(axis=1)
print("Outliers detected by Z score method : \n", outliers_z, "\n")
# 'outliers_z' contains the boolean values of rows containing outliers

# Tukey's Fences method for Outlier Detection
Q1 = merged_data.quantile(0.25)
Q3 = merged_data.quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers_tukey = ((merged_data < lower_bound) | (merged_data > upper_bound)).any(axis=1)
print("Outliers detected by Tukey's Fences method : \n", outliers_tukey)
# 'outliers_tukey' contains boolean values indicating whether each row contains outliers or not
```

OUTLIER DETECTION

```python
plt.figure(figsize=(15, 10))
for i, col in enumerate(merged_data.columns):
    plt.subplot(4, 4, i + 1)
    sns.boxplot(y=merged_data[col], color='skyblue', orient='v')
    plt.title(col)
plt.tight_layout()   # Univariate Outlier Detection
plt.show()

# Multivariate Outlier Detection (PCA)
from sklearn.decomposition import PCA

# Fit PCA and transform the data
pca = PCA(n_components=2)
pca_data = pca.fit_transform(merged_data)

# Plot the PCA results
plt.figure(figsize=(8, 6))
plt.scatter(pca_data[:, 0], pca_data[:, 1], c='skyblue', alpha=0.5)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Boston Housing Features')
plt.show()
```

REMOVING OUTLIER

```python
dt_rmove = merged_data[~outliers_z]

#print
removed_outliers = merged_data[outliers_z]
print("count of data remove:", removed_outliers.shape[0])
```
